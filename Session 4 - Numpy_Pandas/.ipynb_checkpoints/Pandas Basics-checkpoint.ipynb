# IMDB Dataset

```python
%%timeit
movies_df=pd.read_csv('IMDB-Movie-Data.csv',index_col='Title')
```

```python
movies_df.head()
```

```python
movies_df.tail(2)
```

```python
#get information about dataset
movies_df.info()
```

```python
# find number of rows and number of columns
movies_df.shape
```

```python
#Handling duplicates
#append same data and assign to temp_df dataframe
temp_df=movies_df.append(movies_df)
temp_df.shape
```

```python
#remove duplicates
temp_df=temp_df.drop_duplicates()
temp_df.shape
```

# ##Column cleanup

```python
movies_df.columns #get column details from dataframe
```

```python
#Rename the column
movies_df.rename(columns={
    'Runtime (Minutes)':'Runtime',
    'Revenue (Millions)':'Revenue'
},inplace=True)
movies_df.columns
```

```python
movies_df.columns=['rank', 'genre', 'description', 'director', 'actors', 'year', 'runtime',
       'rating', 'votes', 'revenue', 'metascore']
movies_df.columns
#by this way we can change column names to lower case
```

```python
movies_df.head()
```

```python
#change to lower case using list comprehension
movies_df.columns=[col.lower() for col in movies_df]
movies_df.columns
```

### How to work with missing values

```python
#identify null values
movies_df.info()
```

```python
movies_df.isnull()
```

```python
movies_df.isnull().sum()  #returns count of null values in each of columns
```

```python
#dropping null values
temp_df=movies_df.dropna()
temp_df.info()
```

```python
temp_col_df=movies_df.dropna(axis=1) #this will drop the entire column having null values
temp_col_df.shape
```

```python
temp_col_df.info()
```

```python
movies_df.shape
```

### Imputation

```python
#get revenue details in new dataframe
revenue=movies_df['revenue']
revenue.head()
```

```python
#get mean of revenue
revenue_mean=revenue.mean()
revenue_mean
```

```python
#Impute this mean in the null values for revenue
revenue.fillna(revenue_mean,inplace=True)
```

```python
movies_df.isnull().sum()
```

#### Applying functions

```python
def rating_function(x):
    if x>=8.0:
        return "good"
    else:
        return "bad"
```

```python
movies_df['rating_category']=movies_df['rating'].apply(rating_function)
movies_df.head()
```
