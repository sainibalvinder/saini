## Step 0 : Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Step 1: Import Dataset

```python
training_set=pd.read_csv('train.csv')
```

```python
training_set.head()
```

```python
training_set.tail()
```

## Step 2: Explore/Visualize Dataset

```python
survived=training_set[training_set['Survived']==1]
no_survived=training_set[training_set['Survived']==0]
```

```python
#Count the survived and deceased
print('Total =',len(training_set))

print("Survived passengers =",len(survived))
print('% Survived =',len(survived)/len(training_set)*100.0,"%")

print("Did not survived passengers =",len(no_survived))
print("% Did not survived =",len(no_survived)/len(training_set)*100,"%")
```

```python
# Number of people survived based on their class
plt.figure(figsize=[6,12])
plt.subplot(211)
sns.countplot(x='Pclass',data=training_set)
plt.subplot(212)
sns.countplot(x='Pclass',hue='Survived',data=training_set)
```

```python
#Survival rate based on sibling status
plt.figure(figsize=[6,12])
plt.subplot(211)
sns.countplot(x='SibSp',data=training_set)
plt.subplot(212)
sns.countplot(x='SibSp',hue='Survived',data=training_set)
```

```python
#Survival based on Parch status (how many parents onboard)
plt.figure(figsize=[6,12])
plt.subplot(211)
sns.countplot(x='Parch',data=training_set)
plt.subplot(212)
sns.countplot(x='Parch',hue='Survived',data=training_set)
```

```python
# Survival based on port passengers emparked from (C=Cherbourg, Q=Queenstown, S=Southampton)
plt.figure(figsize=[6,12])
plt.subplot(211)
sns.countplot(x='Embarked',data=training_set)
plt.subplot(212)
sns.countplot(x='Embarked',hue='Survived',data=training_set)
```

```python
# Survival based on sex
plt.figure(figsize=[6,12])
plt.subplot(211)
sns.countplot(x='Sex',data=training_set)
plt.subplot(212)
sns.countplot(x='Sex',hue='Survived',data=training_set)
```

```python
#Bar chart to indicate number of people survived based on age
plt.figure(figsize=(40,30))
sns.countplot(x='Age',hue='Survived',data=training_set)
```
